```python
# Agrupar los datos por marca y modelo, y calcular el precio promedio y el kilometraje promedio para cada combinación.
df=pd.read_csv('C:/afiles/python/Venta_Autos.csv')
df.drop(['saledate'],axis=1, inplace=True)
df1=df.groupby(by=['make','model']).agg({
    "sellingprice": "mean",
    "odometer": "mean"
}).reset_index()
df1.head(500000)
```

```python
# 4.**Tidy Data**
# - Transformar el DataFrame a un formato "tidy", donde cada variable tiene su propia columna y cada observación su propia fila.

df['kilometraje']= pd.qcut(df['odometer'],3,labels=['bajo','medio','alto'])
df
pivot_table = df.pivot_table(values='sellingprice', index=['make','model', 'year'], columns='kilometraje', aggfunc='mean')
pivot_table.head(500000)
```

```python
display(pivot_table) #Dataframe original con indices multi nivel indicando precio promedio segun percentil de kilometraje alto-medio-alto
display(pivot_table.reset_index())# Dataframe con los indices "reseteados" se llevan todos al mimsmo nivel

tidy_df = pivot_table.reset_index().melt(id_vars=['make','model','year'], var_name='kilometraje', value_name='sellingprice') #se reorganiza el dataframe para que la clasificacion de percentil de kilometraje vuelva a ser una unica columna. esto es util en caso de que debamos estructurar datos en forma de tablas para bases de datos.

display(tidy_df)
```

4.  **Tidy Data**
    -   Documentar los pasos necesarios para realizar esta transformación y explicar por qué es importante tener datos en formato "tidy".

### Beneficios de Usar Datos Tidy

1. **Facilita el Análisis:**
   - Los datos tidy son más fáciles de manipular y analizar con herramientas estadísticas y de visualización de datos.

2. **Compatibilidad con Herramientas:**
   - Muchas herramientas de análisis de datos, como Pandas en Python, están diseñadas para trabajar con datos tidy, lo que permite aprovechar al máximo sus capacidades.

3. **Reproducibilidad:**
   - Al tener un formato estándar, es más fácil compartir y reproducir análisis de datos con otros profesionales.

4. **Eficiencia en la Manipulación de Datos:**
   - Transformar, agregar y filtrar datos es más sencillo y eficiente cuando los datos están en un formato tidy.

```python
# 6.6 - ¿Existe alguna correlación entre el valor del mercado de Manheim (mmr) y el precio de venta (sellingprice)?
df = pd.read_csv('C:/afiles/python/Venta_Autos.csv')
df.drop(['saledate'],axis=1, inplace=True)
#df.describe()

df=df.groupby(by=['make','model']).agg({
    "sellingprice": "mean",
    "mmr": "mean",
    "odometer":"mean"
}).reset_index()
df.head(50)


df['SelxKm']=df["sellingprice"]/df["odometer"]
df['mmrxKm']=df["mmr"]/df["odometer"]
df['MmrvsSel']=df["SelxKm"]-df["mmrxKm"]
df.sort_values(by='MmrvsSel',inplace=True,ascending=False)
df.head(10)
```

```python
# 6.7 - ¿Cuál es el volumen total de ventas de vehículos en el dataset?
pd.set_option("styler.format.thousands", ",") 
ventastotales=df['sellingprice'].sum()
res = '{:,.2f}'.format(ventastotales)
print ("el volumen total de ventas es: $", res)
```

```python
# 6.9 - ¿Cuántos vehículos de cada marca debería tener en inventario una concesionaria?
import pandas as pd
import datetime
 
df = pd.DataFrame
df = pd.read_csv('C:/afiles/python/Venta_Autos.csv')
 
df['saledate'] = pd.to_datetime(df['saledate'],utc=True, errors='coerce')
 
df = df.dropna(subset=['saledate'])
 
dfvxd = df.groupby(['make', df['saledate'].dt.date]).size().reset_index(name='Cantidad')
 
dfpxmarca = dfvxd.groupby('make')['Cantidad'].mean().reset_index(name='PromedioDiario')
 
periodo_stock = 30
 
dfpxmarca['CantidadNecesaria'] = dfpxmarca['PromedioDiario'] * periodo_stock
 
dfpxmarca['PromedioDiario']=dfpxmarca['PromedioDiario'].map("{:,.0f} unidades".format)
dfpxmarca['CantidadNecesaria']=dfpxmarca['CantidadNecesaria'].map("{:,.0f} unidades".format)
print("Promedio diario de ventas por marca:")
display(dfpxmarca[['make', 'PromedioDiario']])
 
print("\nCantidad de autos necesarios para un período de stock de 30 días:")
display(dfpxmarca[['make', 'CantidadNecesaria']])
```

```python
#10 - ¿Qué proporción de vehículos con diferentes tipos de carrocería debería tener una concesionaria? 
df = pd.DataFrame
df = pd.read_csv('C:/afiles/python/Venta_Autos.csv')
 
df['saledate'] = pd.to_datetime(df['saledate'],utc=True, errors='coerce')
 
df = df.dropna(subset=['saledate'])
 
dfvxd = df.groupby(['seller','body', df['saledate'].dt.date]).size().reset_index(name='Cantidad')
 
dfpxmarca = dfvxd.groupby(['seller','body'])['Cantidad'].mean().reset_index(name='PromedioDiario')
 
periodo_stock = 30
 
dfpxmarca['CantidadNecesaria'] = dfpxmarca['PromedioDiario'] * periodo_stock
 
dfpxmarca['PromedioDiario']=dfpxmarca['PromedioDiario'].map("{:,.0f} unidades".format)
dfpxmarca['CantidadNecesaria']=dfpxmarca['CantidadNecesaria'].map("{:,.0f} unidades".format)
print("Promedio diario de ventas por consecionaria y carroceria:")
display(dfpxmarca[['seller','body', 'PromedioDiario']])
 
print("\nCantidad de autos necesarios para un período de stock de 30 días:")
display(dfpxmarca[['seller','body', 'CantidadNecesaria']])
```

```python
# 6.11 - ¿Qué cantidad de vehículos de cada año de fabricación debería tener en inventario?
 
df = pd.DataFrame
df = pd.read_csv('C:/afiles/python/Venta_Autos.csv')
infer_datetime_format=True

df['saledate'] = pd.to_datetime(df['saledate'],utc=True, errors='coerce')
 
df = df.dropna(subset=['saledate'])
 
dfvxd = df.groupby(['make','year', df['saledate'].dt.date]).size().reset_index(name='Cantidad')
 
dfpxmarca = dfvxd.groupby(['make','year'])['Cantidad'].mean().reset_index(name='PromedioDiario')
 
periodo_stock = 30
 
dfpxmarca['CantidadNecesaria'] = dfpxmarca['PromedioDiario'] * periodo_stock
 
dfpxmarca['PromedioDiario']=dfpxmarca['PromedioDiario'].map("{:,.0f} unidades".format)
dfpxmarca['CantidadNecesaria']=dfpxmarca['CantidadNecesaria'].map("{:,.0f} unidades".format)
print("Promedio diario de ventas por marca y año:")
display(dfpxmarca[['make','year', 'PromedioDiario']])
 
print("\nCantidad de autos necesarios para un período de stock de 30 días:")
display(dfpxmarca[['make','year', 'CantidadNecesaria']])
```
